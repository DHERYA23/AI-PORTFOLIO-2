```python
import pandas as pd
df = pd.read_csv('data.csv')
```

```python
df.head()
```

```python
df["Date"]
```

```python
df_new = df.drop(1)
```

```python
import matplotlib.pyplot as myplot
x = df_new.Date
y = df_new.max_temperature
myplot.scatter(x,y)
myplot.xticks(rotation = 90)
myplot.title("Mean Temperature")
myplot.xlabel("Date", fontsize = 20)
myplot.ylabel("Maximum Tempertaure", fontsize = 20)
myplot.show()
```
